```python
from time import time
from time import sleep
from random import randint
from IPython.core.display import clear_output
from warnings import warn
from bs4 import BeautifulSoup
import requests
```

```python
url = ' https://www.space.com/news'
```

```python
response = requests.get(url)
```

```python
if (response.ok):
    data = response.text
    print(data)
```

```python
## using BeautifulSoup to prettify the data
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
print(soup.prettify())
```

```python
raw_stories = soup.select('.content')
print(raw_stories)
```

```python
def process_page(soup,results):
    raw_stories = soup.select('.content')
    for story in raw_stories:
        headline = story.select_one('.content > header > h3').get_text()
        author = story.select_one('.content > header > p > span').get_text().replace('\nBy\n\n','').replace('\n','')
        synopsis = story.select_one('.content > p').get_text().replace('\n','')
        date_time = story.select_one('.content > header > p > time')["data-published-date"]
        story = {
                'headline': headline,
                'author':author,
                'synopsis': synopsis,
                'date_Time': date_time
        }   
        
        results.append(story)
        
start_time = time()
request_count = 0
results = []
has_next_page = True
MAX_REQUESTS = 100
page_number = 1
url = 'https://www.space.com/news'

headers = {'user-agent':'newscraper-school project(my email:sajithgowthaman@gmail.com)'}
while has_next_page and request_count <= MAX_REQUESTS:
    clear_output(wait = True)   
    query = {'pg': page_number}
    response = requests.get(url, params=query, headers=headers)
    if(response.ok):
        data = response.text
        soup = BeautifulSoup(data, 'html.parser')
        process_page(soup, results)
        next_button = soup.select('.listings-next')
        has_next_page = len(next_button) > 0
        print(has_next_page)
    else:
        warn('Request #: {}, Failed with status code: {}'.format(request_count, response.status_code))
    request_count += 1
    sleep(randint(1,3))     
    elapsed_time = time() - start_time
    print('Requests: {}, Frequency: {} requests/s, {} story processed.'.format(request_count, request_count/elapsed_time, len(results)))
    page_number += 1
print('Scraping complete')
print('Requests: {}, Frequency: {} requests/s, {} story processed.'.format(request_count, request_count/elapsed_time, len(results)))
```
