```python
## Creating a list to append the values of the record for all the web pages.
records = []
while True:
        if "next" in data["info"].keys():
            records.append(data["records"][0])
            next_url = data["info"]["next"]
            print(next_url)
            data = requests.get(next_url).json()
            
        else:
            records.append(data["records"][0])
            break 
```

```python
print(len(records))
print(records)
```

```python
## Defining a function to return the display name, birthplace and culture from raw data.
def process_data(records):
    return [{
      'displayname' : record['displayname'],
      'birthplace': record['birthplace'],
      'culture': record['culture']   
        } for record in records ]
```

```python
output = process_data(records)
output
```
